### 1. Simulating synthetic data

```python
# The Generative Model

# Number of observations, n = 750 
# Number of variables, p = 4
# np.random.seed(0)
N = 750;p = 4 # training
Ntest = 250;
X = [[np.random.uniform(0,9)/10 for i in range(p)] for j in range(N)]
X = np.matrix(X)
# simulating the data
Xtest = [[np.random.uniform(0,9)/10 for i in range(p)] for j in range(Ntest)]
Xtest = np.matrix(Xtest)
print(X.shape)
print(Xtest.shape)
```

```python
X = pd.DataFrame(X)
Xtest = pd.DataFrame(Xtest)
X.columns = [ 'x' + str(i+1) for i in range(p)]
Xtest.columns = [ 'x' + str(i+1) for i in range(p)]
X.head()
```

```python
rulelist =  [
  ' (x3 <= 0.6) & (x2 < 0.5) ', 
  ' (x2 > 0.2) & (x4 >= 0.3) ',
  ' (x3 > 0.4) &  (x2 < 0.8) '
]
```

```python
part = partitiondata(rulelist, X, test = 1, minth = 50)
parttest = partitiondata(rulelist, Xtest, test = 1, minth = 10)
```

```python
def alphaf(x):
    f = -4
    for i in range(2):
        f = f + np.exp(-20*((x[i] - 0.5)**2))
        
    f = f + np.sin(2*np.pi*(x[2]**2))
    f = f + np.sin(2*np.pi*(x[3]**2))
    return 24*np.exp(f/4)+15

def betaf(x):
    f = 0
    f = f + np.sin(2*np.pi*(x[0])**2) + np.sin(2*np.pi*(x[1]**2))
    f = f + np.cos(2*np.pi*(x[2]**2)) + np.cos(2*np.pi*(x[3]**2))
    return 12*np.exp(f/12)+8

def alphaf2(x):
    f = -2
    for i in range(2):
        f = f + np.exp(-10*((x[i] - 0.5)**2))
        
    f = f + np.sin(2*np.pi*(x[2]**2))
    f = f + np.sin(2*np.pi*(x[3]**2))
    return 2*np.exp(f)+2

def betaf2(x):
    f = +2
    f = f + np.sin(2*np.pi*(x[0])**2) + np.sin(2*np.pi*(x[1]**2))
    f = f + np.cos(2*np.pi*(x[2]**2)) + np.cos(2*np.pi*(x[3]**2))
    return 8*np.exp(f/12) - 5

def alphaf3(x):
    f = -4
    for i in range(0,2):
        f = f + np.exp(5*((x[i] - 0.25)**4))
        
    f = f + np.sin(1*np.pi*(x[2]**1))
    f = f + np.sin(1*np.pi*(x[3]**1))
    return 0.02*np.exp(f/16)

def betaf3(x):
    f = 0
    f = f + np.sin(2*np.pi*(x[2])**2) + np.sin(2*np.pi*(x[3]**2))
    f = f + np.cos(2*np.pi*(x[0]**2)) + np.cos(2*np.pi*(x[1]**2))
    return 0.75*np.exp(f/16) + 0.1

def alphaf4(x):
    f = -2
    for i in range(2):
        f = f + np.exp(-10*((x[i] - 0.5)**2))
        
    f = f + np.sin(2*np.pi*(x[2]**2))
    f = f + np.sin(2*np.pi*(x[3]**2))
    return 12*np.exp(f/4)

def betaf4(x):
    f = 0
    f = f + np.sin(2*np.pi*(x[0])**2) + np.sin(2*np.pi*(x[1]**2))
    f = f + np.cos(2*np.pi*(x[2]**2)) + np.cos(2*np.pi*(x[3]**2))
    return 12*np.exp(f/16)+2
```

```python
alpha = np.array([0. for i in range(N)])
beta = np.array([0. for i in range(N)])
alphatest = np.array([0. for i in range(Ntest)])
betatest = np.array([0. for i in range(Ntest)])
for i in range(part['numparts']):

# for i in range(1):
    idx = part[i]['conditional'].index[part[i]['conditional']]
    idxtest = parttest[i]['conditional'].index[parttest[i]['conditional']]

    Xidx = X[part[i]['conditional']]
    Xidxtest = Xtest[parttest[i]['conditional']]
    mbeta = 1 #2 #*np.random.uniform(1,2)
    malpha = 1 #10*np.random.uniform(0,3)
    
    if i == 0:
        alphaidx = [alphaf4(x) for x in np.array(Xidx)]
        betaidx = [betaf4(x) for x in np.array(Xidx)]
        
        alphaidxtest = [alphaf4(x) for x in np.array(Xidxtest)]
        betaidxtest = [betaf4(x) for x in np.array(Xidxtest)]
    if i == 1:
        alphaidx = [alphaf2(x) for x in np.array(Xidx)]
        betaidx = [betaf2(x) for x in np.array(Xidx)]
        
        alphaidxtest = [alphaf2(x) for x in np.array(Xidxtest)]
        betaidxtest = [betaf2(x) for x in np.array(Xidxtest)]
    if i == 2:
        alphaidx = [alphaf3(x) for x in np.array(Xidx)]
        betaidx = [betaf3(x) for x in np.array(Xidx)]
        
        alphaidxtest = [alphaf3(x) for x in np.array(Xidxtest)]
        betaidxtest = [betaf3(x) for x in np.array(Xidxtest)]
    if i == 3:
        alphaidx = [alphaf(x) for x in np.array(Xidx)]
        betaidx = [betaf(x) for x in np.array(Xidx)]
        
        alphaidxtest = [alphaf(x) for x in np.array(Xidxtest)]
        betaidxtest = [betaf(x) for x in np.array(Xidxtest)]
    if i == 4:
        alphaidx = [alphaf5(x) for x in np.array(Xidx)]
        betaidx = [betaf5(x) for x in np.array(Xidx)]
        
        alphaidxtest = [alphaf5(x) for x in np.array(Xidxtest)]
        betaidxtest = [betaf5(x) for x in np.array(Xidxtest)]
        
    alpha[idx] = alphaidx
    beta[idx] = betaidx
    
    alphatest[idxtest] = alphaidxtest
    betatest[idxtest] = betaidxtest
```

```python
np.random.seed(1)
time1 = [fisk.rvs(c = b, loc=0, scale = a, size=1, random_state=None) for b,a in zip(beta,alpha)]
np.random.seed(1)
cens = [np.random.binomial(1,0.8) for i in range(N)]
np.random.seed(1)
rho = [np.random.uniform(0.6,1) for i in range(N)]
time = [t*r if c == 0 else t for t,r,c in zip(time1,rho,cens)] 

timetest1 = [fisk.rvs(c = b, loc=0, scale = a, size=1, random_state=None) for b,a in zip(betatest,alphatest)]
censtest = [np.random.binomial(1,0.8) for i in range(Ntest)]
rhotest = [np.random.uniform(0.6,1) for i in range(Ntest)]
timetest = [t*r if c == 0 else t for t,r,c in zip(timetest1,rhotest,censtest)] 

ts = np.stack((np.array(time),np.array(cens).reshape(-1,1)), axis = 1).reshape(-1,2)
D = np.concatenate((ts,X), axis = 1)
D = pd.DataFrame(D)

ts = np.stack((np.array(timetest),np.array(censtest).reshape(-1,1)), axis = 1).reshape(-1,2)
Dtest = np.concatenate((ts,Xtest), axis = 1)
Dtest = pd.DataFrame(Dtest)

D.columns = ['time', 'status'] +[ 'x' + str(i+1) for i in range(p)]
D.status = pd.get_dummies(D.status)[1]

Dtest.columns = ['time', 'status'] +[ 'x' + str(i+1) for i in range(p)]
Dtest.status = pd.get_dummies(Dtest.status)[1]

time = np.array(D.time).reshape(-1,1)
cens = np.array(D.status).reshape(-1,1)
X = D.loc[:,~D.columns.isin(['time','status'])]

timetest = np.array(Dtest.time).reshape(-1,1)
censtest = np.array(Dtest.status).reshape(-1,1)
Xtest = Dtest.loc[:,~Dtest.columns.isin(['time','status'])]

print(max(alpha))
print(min(alpha))
print(max(beta))
print(min(beta))

kmf = KaplanMeierFitter()
kmf.fit(time, cens, label = 'total')
kmf.plot(ci_show = True)
plt.xlabel('Time')
plt.legend()
plt.show()
data = D.copy()
```

## FULL GP SURV MODEL

```python
# GP TOTAL
## INFERENCE
input_dim = X.shape[1]
print(input_dim)
Y_metadata = {'censored': 1-cens}
kprior = GPy.kern.RBF(input_dim = X.shape[1], name = 'f_rbf', ARD = True)
kprior.variance.set_prior(GPy.priors.Gamma.from_EV(1,10))
kprior.lengthscale.set_prior(GPy.priors.Gamma.from_EV(1,10))
lik = GPy.likelihoods.LogLogistic()
lik.r_log_shape.set_prior(GPy.priors.Uniform(0,50))
laplace_inf = GPy.inference.latent_function_inference.Laplace()
## INFERENCE # ONE MODEL
m01 = GPy.core.GP(X=X, Y=time, likelihood=lik,
                 inference_method=laplace_inf, kernel=kprior,
                 Y_metadata = Y_metadata)
print(m01)
print(m01.optimize('bfgs', messages=1, max_iters=1300))
ax = plot_St(m01, np.array(X), label = '$\it{if}$' + ' ' + '$r_1$', ax = None, conf_int = True)
```

```python
prediction_score01 = sum(m01.log_predictive_density_sampling(np.array(Xtest),timetest,Y_metadata = {'censored' : 1-censtest}))
print('The prediction score of full GP model is: {}'.format(prediction_score01))
loglikelihood01 = m01.log_likelihood() # -699.93794143
print('The loglikelihood of full GP model is: {}'.format(loglikelihood01))
```
